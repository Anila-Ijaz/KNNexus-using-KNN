```python
import numpy as np
import pandas as pd
import sklearn.datasets      
from KNNClassifier_Algorithm import KNNClassifier      # importing the custom KNN class
import sklearn.model_selection       # for train test split
from sklearn import metrics as m       # for accuracy calculation
from sklearn.metrics import confusion_matrix
```

```python
dataset = sklearn.datasets.load_iris() #Iris Dataset
X = dataset.data # extracting data points
y = dataset.target # extracting labels
```

```python
iris_data= pd.DataFrame(data= np.c_[dataset['data'], dataset['target']],
                     columns= dataset['feature_names'] + ['target'])
iris_data.describe()
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.30, train_size=0.70)
```

Custom KNN Classifier

```python
k = KNNClassifier(k_neighbors=4) # creating an object of custom KNN class with k=4
k.fit(X_train, y_train); # a custom function to fit the TR data is called.
CustomKNNClassifierPrediction = k.predict(X_test); # a custom function to predict the labels of TE data is called.
```

```python
m.accuracy_score(y_test, CustomKNNClassifierPrediction) # calculating accuracy of custom KNN classifier
```

```python
conf_matrix_Custom_KNN_Dataset1 = confusion_matrix(y_test, CustomKNNClassifierPrediction)
df = pd.DataFrame(conf_matrix_Custom_KNN_Dataset1, index=["Actual_0", "Actual_1", "Actual_2"], columns=["Pred_0", "Pred_1", "Pred_2"])
df.to_csv("Dataset1_Classifier1.csv")
```

SKLearn KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
neighbor = KNeighborsClassifier(n_neighbors=3) # making an object of the sklearn KNNClassifier class
neighbor.fit(X_train, y_train) # training the object with TR 
prediction = neighbor.predict(X_test)
m.accuracy_score(y_test, prediction) # calculating accuracy of sklearn KNN classifier
```

```python
conf_matrix_SkLearn_KNN_Dataset1 = confusion_matrix(y_test, prediction)
df1 = pd.DataFrame(conf_matrix_SkLearn_KNN_Dataset1, index=["Actual_0", "Actual_1", "Actual_2"], columns=["Pred_0", "Pred_1", "Pred_2"])
df1.to_csv("Dataset1_Classifier2.csv")
```

 SKLearn Tree Classifier

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier() # making an object of the DecisionTreeClassifier class
clf = clf.fit(X_train, y_train) # training the object with TR
predictionDecisionTrees = clf.predict(X_test); # predicting the labels of TE data
m.accuracy_score(y_test, predictionDecisionTrees) # calculating accuracy of decision tree classifier
```

```python
conf_matrix_SkLearn_DecisionTrees_Dataset1 = confusion_matrix(y_test, predictionDecisionTrees)
df2 = pd.DataFrame(conf_matrix_SkLearn_DecisionTrees_Dataset1, index=["Actual_0", "Actual_1", "Actual_2"], columns=["Pred_0", "Pred_1", "Pred_2"])
df2.to_csv("Dataset1_Classifier3.csv")
```
